```python
import spacy
import textacy

length = len(train['full_text'])

def clean(train):
  length = len(train['full_text'])
  train['sentences'] =[0 for i in range(length)]
  nlp = spacy.load('en_core_web_sm')  
  for i in range(length):
    doc = nlp(train['full_text'][i])
    sentences = list(doc.sents)
    train['sentences'][i] = sentences
  return train 
```

```python
train = clean(data)
```

```python
train
```

```python
sen_value = {'adjectives':'ADJ','adposition':'ADP','adverb':'ADV','conjuction':'CONJ','coordinating_conjunction':'CCONJ','determiner':'DET','interjection':'INTJ',"noun":'NOUN','numeral':'NUM','particle':'PART','prnoun':'PRON','proper_noun':'PROPN','punctuation':'PUNCT','subordination_conjuction':'SCONJ','symbol':'SYM','verbs':'VERB','space':'SPACE'}
sen_value_len = list()
for i in sen_value.keys():
  sen_value_len.append(i+'_len')
```

```python
def extract(train):
  length = len(train['full_text'])
  for i in sen_value.keys():
    train[i] =[ [] for i in range(length)]
  for i in sen_value_len:
    train[i] = [0 for i in range(length)]
  for i in range(length):
    sen_list = train['sentences'][i]
    for j in sen_value.keys():
      for k in sen_list:
        pattern = [{'POS':sen_value[j]}]
        train[j][i] += list(textacy.extract.matches.token_matches(k,patterns=pattern))
  for i in sen_value.keys():
    l = 0
    for j in range(length):
      train[i+'_len'][j] = len(train[i][j])

  return train
```

```python
train = extract(train)
train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

X = train[sen_value_len]

for i in ['cohesion','syntax','vocabulary','phraseology','grammar','conventions']:
  train[i] = list(map(float,train[i]))

  y = train['cohesion']
  train_x,val_x,train_y,val_y = train_test_split(X,y)

  model = RandomForestRegressor()
  model.fit(train_x,train_y)
  predictions = list(map(float,model.predict(test_X)))
  test[i] = predictions
```
